```python
#import dependencies
from citipy import citipy
import pandas as pd
import numpy as np
import requests
import json
import openweathermapy.core as owm
import random
import matplotlib.pyplot as plt

from weather_config import api_key
```

```python
#get random 500 random latitude and longitude points

#create a list
latitude =[]
longitude=[]

#loop 700 times
for x in range(2000):
    
    #getting latitude
    #latitude range is -90 to 90
    lat=random.randrange(-90,90)+random.random()
    latitude.append(lat)
    
    #getting longitude
    #longitude range is -180 to 180
    lon=random.randrange(-180,180)+random.random()
    longitude.append(lon)

# print(f"lat: {latitude}")
# print(f"lon: {longitude}")


#iterate to get city names for nearest ciry
CityName = []

for x in range(2000):
    city=citipy.nearest_city(latitude[x], longitude[x])
    CityName.append(city.city_name)
    
#create a dataframe
city_list = pd.DataFrame ({"latitude": latitude, "longitude": longitude,"City Name": CityName})
city_list.head(10)
print(len(city_list))

```

```python
#drop duplicate cities
city_list=city_list.drop_duplicates("City Name")
#reset index
city_list=city_list.reset_index()
city_list.head(50)
#size of dataframe
print(len(city_list))
```

```python
#getting Temperature, Humidity, Cloudiness and Windspeed for cities
#creating settings
settings = {"units": "metric", "appid": api_key}
ProcessingRecord=1

#iterate through dataframe and update column values
for index,row in city_list.iterrows():
    try:
        print(f"Processing Record {ProcessingRecord} | {row['City Name']}")
        print(f"{owm.BASE_URL}weather?q={row['City Name']} ")
        ProcessingRecord = ProcessingRecord + 1 
        data=owm.get_current(f"{row['City Name']}",**settings)
        city_list.set_value(index,"Temperature",data["main"]["temp"])
        city_list.set_value(index,"Humidity",data["main"]["humidity"])
        city_list.set_value(index,"Cloudiness",data["clouds"]["all"]) 
        city_list.set_value(index,"WindSpeed",data["wind"]["speed"])
    #if there is an error, print the city name and skip it
    except:
        print(f"{row['City Name']} does not exist")
        continue
              
```

```python
city_list.head()
```

```python
#drop rows where no information was pulled
city_list=city_list.dropna()
#reset index
city_list = city_list.reset_index(drop=True)
#pull only columns that I want
city_list = city_list[["City Name", "latitude", "longitude", "Temperature","Humidity","Cloudiness", "WindSpeed"]]
city_list["Temperature"]= (9/5)*city_list["Temperature"]+32
```
